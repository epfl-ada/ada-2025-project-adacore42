```python
def extract_wiki_titles(entity_list):
    return [
        url.split("wikipedia.org/wiki/")[1]
        for url in entity_list
        if "wikipedia.org/wiki/" in url
    ]

wiki_titles = dataC2["entities"].apply(extract_wiki_titles)
print(wiki_titles)
```
