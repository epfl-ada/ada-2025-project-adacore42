```python
import statsmodels.formula.api as smf
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
dataAm_rf = pd.read_csv("/Users/todorovkatia/Documents/EPFL/MA3/ADA/Projet/code_perso/dataAm_rf.csv")
dataAm_rf_clean = dataAm_rf[dataAm_rf["num_words"]<=80]
dataAm_rf_clean
```

```python
sns.boxplot(dataAm_rf_clean["num_repeats"])

dataAm_rf_clean.loc[:, "category"] = np.where(dataAm_rf_clean['funny_score'] > 0, '1', '0') #1 for funny, 0 for unfunny
dataAm_rf_clean.columns
dataAm_rf_clean
```

```python
# how we standardize the countinuous variables
df = dataAm_rf_clean.copy()
df['num_words'] = (df['num_words'] - df['num_words'].mean())/df['num_words'].std()
df['polarity'] = (df['polarity'] - df['polarity'].mean())/df['polarity'].std()
df['subjectivity'] = (df['subjectivity'] - df['subjectivity'].mean())/df['subjectivity'].std()
df['num_proper_nouns'] = (df['num_proper_nouns'] - df['num_proper_nouns'].mean())/df['num_proper_nouns'].std()
df['num_verbs'] = (df['num_verbs'] - df['num_verbs'].mean())/df['num_verbs'].std()
df['num_repeats'] = (df['num_repeats'] - df['num_repeats'].mean())/df['num_repeats'].std()
df['num_punct'] = (df['num_punct'] - df['num_punct'].mean())/df['num_punct'].std()
```

## Ordinary least squares linear regression

```python
dataAm_rf_clean.dtypes
```

```python
dataAm_rf_clean["category"] = dataAm_rf_clean["category"].astype(int)
```

```python
mod = smf.ols(formula='category~ num_words + num_verbs', data=dataAm_rf_clean)
```

```python
res = mod.fit()
print(res.summary())
```

## Logistic regression

```python
mod = smf.logit(formula='category ~  num_words + polarity + subjectivity + \
                        num_proper_nouns + num_verbs + num_repeats + \
                        num_punct ', data=df)
res = mod.fit()
print(res.summary())
```

```python
df_majority = df[df['category'] == 0]
df_minority = df[df['category'] == 1]
df_majority_downsampled = df_majority.sample(n=len(df_minority), random_state=42)
df_balanced = pd.concat([df_majority_downsampled, df_minority]).sample(frac=1, random_state=42)
```
