### Style and structure

- Length of the captions: is there a correlation between length and votes ?

More precisely, does this correlation, if it exists, evolve over time ?

```python
dataA1 = dataA.copy()
```

```python
def lenCaptionVsVotes(dataA, dataA_startID, idx):

    # Visualisation of the relationship between caption length and vote for a particular contest.

    df = dataA[idx].copy()
    df['caption_length'] = df['caption'].str.len()

    sns.scatterplot(data=df, x='caption_length', y='votes', alpha=0.6)
    sns.regplot(data=df, x='caption_length', y='votes', scatter=False, color='red')
    plt.title(f"Contest {idx+dataA_startID} : Votes vs Length")
    plt.show()
```

```python
# Let's compute the correlation for each contest

correlations = []

dataA_startID = DataPreparation.dataA_startID

for i, df in enumerate(dataA1):

    df['caption_length'] = df['caption'].str.len()
    corr = df['caption_length'].corr(df['votes'])
    
    correlations.append({
        'contest_id': dataA_startID + i, 
        'corr_length_votes': corr,
        'n_captions': len(df)
    })

# Put it all in a df
corrs_df = pd.DataFrame(correlations)
print(corrs_df.head())

# Mean correlation over all of the contests
mean_corr = corrs_df['corr_length_votes'].mean()
print(f"Mean correlation (length ↔ votes) = {mean_corr:.3f}")

# Visualisation
plt.figure(figsize=(10,5))
sns.lineplot(data=corrs_df, x='contest_id', y='corr_length_votes', marker='o')
plt.title("Correlation between caption length and votes over contests")
plt.xlabel("Contest ID")
plt.ylabel("Correlation (r)")
plt.show()
```
