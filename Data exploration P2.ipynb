**Length of the captions**: is there a correlation between length and votes ?

More precisely, does this correlation, if it exists, evolve over time ?

```python
dataA1 = dataA.copy()
```

```python
def lenCaptionVsVotes(dataA, dataA_startID, idx):

    # Visualisation of the relationship between caption length and vote for a particular contest.

    df = dataA[idx].copy()
    df['caption_length'] = df['caption'].str.len()

    sns.scatterplot(data=df, x='caption_length', y='votes', alpha=0.6)
    sns.regplot(data=df, x='caption_length', y='votes', scatter=False, color='red')
    plt.title(f"Contest {idx+dataA_startID} : Votes vs Length")
    plt.show()
```

```python
# Let's compute the correlation for each contest

correlations = []

dataA_startID = DataPreparation.dataA_startID

for i, df in enumerate(dataA1):

    df['caption_length'] = df['caption'].str.len()
    corr = df['caption_length'].corr(df['votes'])
    
    correlations.append({
        'contest_id': dataA_startID + i, 
        'corr_length_votes': corr,
        'n_captions': len(df)
    })

# Put it all in a df
corrs_df = pd.DataFrame(correlations)
print(corrs_df.head())

# Mean correlation over all of the contests
mean_corr = corrs_df['corr_length_votes'].mean()
print(f"Mean correlation (length ↔ votes) = {mean_corr:.3f}")

# Visualisation
plt.figure(figsize=(10,5))
sns.lineplot(data=corrs_df, x='contest_id', y='corr_length_votes', marker='o')
plt.title("Correlation between caption length and votes over contests")
plt.xlabel("Contest ID")
plt.ylabel("Correlation (r)")
plt.show()
```

```python
lenCaptionVsVotes(dataA1, dataA_startID, 102)
```

**Conclusion**

It seems that there is no clear relationship between the caption length and the number of votes.

```python
def findPunctuation(df, punct_type = 'has_exclamation', plot = False):

    # find it the caption contains the punctuation
    df['has_exclamation'] = df['caption'].str.contains('!', regex=False)
    df['has_quotes'] = df['caption'].str.contains(r'["“”]', regex=True)
    df['has_ellipsis'] = df['caption'].str.contains(r'\.\.\.', regex=True)

    # count the occurences
    df['exclamation_count'] = df['caption'].str.count('!')
    df['quotes_count'] = df['caption'].str.count(r'["“”]')
    df['ellipsis_count'] = df['caption'].str.count(r'\.\.\.')


    if plot:
        sns.histplot(data=df, x='votes', hue=punct_type, bins=30, kde=True, element='step')
        plt.title(f'Votes distribution for captions that {punct_type} or not')
        plt.show()

    return df

df_ = findPunctuation(dataA2[0], punct_type='has_ellipsis', plot = True)
```

```python
def presenceOfPunctuation(df):

    df_punc = findPunctuation(df)

    excl = (
        df_punc.groupby('contest_id')['has_exclamation']
        .mean()
        .reset_index(name = 'perc_with_exclamation')
    )
    excl['perc_with_exclamation'] = excl['perc_with_exclamation'].astype(float)

    quotes = (
        df_punc.groupby('contest_id')['has_quotes']
        .mean()
        .reset_index(name = 'perc_with_quotes')
    )
    quotes['perc_with_quotes'] = quotes['perc_with_quotes'].astype(float)
    

    ellipsis = (
        df_punc.groupby('contest_id')['has_ellipsis']
        .mean()
        .reset_index(name = 'perc_with_ellipsis')
    )
    ellipsis['perc_with_ellipsis'] = ellipsis['perc_with_ellipsis'].astype(float)

    return excl, quotes, ellipsis

def plot_punctuation_evolution(excl, quotes, ellipsis):
    """
    Plot mean of punctuation presence across contests.
    """
    
    plt.figure(figsize=(10,5))
    plt.plot(excl['contest_id'].astype(float), excl['perc_with_exclamation'], label=f'exclamation', color='tab:blue')
    plt.plot(quotes['contest_id'].astype(float), quotes['perc_with_quotes'], label=f'quotes', color='tab:red')
    plt.plot(ellipsis['contest_id'].astype(float), ellipsis['perc_with_ellipsis'], label=f'ellipsis', color='tab:green')

    plt.xlabel('Contest ID')
    plt.ylabel('% of captions with punctuation')
    plt.ylim(0,0.6)
    plt.title(f'Evolution of punctuation usage across contests')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
excl, quotes, ellipsis = presenceOfPunctuation(dataA_allFiles1)
```

```python
plot_punctuation_evolution(excl, quotes, ellipsis)
```
